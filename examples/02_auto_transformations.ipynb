```python
import json

import pandas as pd
from sklearn.datasets import load_wine
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
from sklearn.pipeline import Pipeline

from beaverfe import BeaverPipeline, auto_feature_pipeline
```

# 1. Get the dataset

```python
data = load_wine()

df = pd.DataFrame(data.data, columns=data.feature_names)
df["target"] = data.target

x, y = df.drop(columns="target"), df["target"]

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42, stratify=y)
```

# 2. Define the model

```python
model = KNeighborsClassifier()
scoring = "accuracy"
direction = "maximize"
```

# 3. Evaluate base model

```python
pipe = Pipeline(steps=[("m", model)])
pipe.fit(x_train, y_train)

base_score = pipe.score(x_test, y_test)
print(f"Baseline accuracy: {base_score:.3f}")
```

# 4. Run auto transformations

```python
transformations = auto_feature_pipeline(x_train, y_train, model, scoring, direction)

# Create transformer based on transformations
transformer = BeaverPipeline(transformations)
```

```python
print(json.dumps(transformations, indent=4))
```

# 5. Evalute

```python
pipe = Pipeline(steps=[("t", transformer), ("m", model)])
pipe.fit(x_train, y_train)

score = pipe.score(x_test, y_test)
print(f"Improved: {base_score:.3f} -> {score:.3f}")
```
